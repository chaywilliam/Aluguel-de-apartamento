```python
import pandas as pd
```

```python
# Leitura do arquivo Excel
aluguel_casa = pd.read_excel("/content/houses_rent.xlsx")
```

```python
# Exibição das 10 primeiras linhas
aluguel_casa.head(10)
```

```python
# Exibição das 10 últimas linhas
aluguel_casa.tail(10)
```

```python
# Renomeando colunas para melhor legibilidade
aluguel_casa = aluguel_casa.rename(columns={"parking spaces":"parking_spaces", "rent amount":"rent_amount", "property tax":"property_tax", "fire insurance":"fire_insurance"})
```

```python
# Exibição das primeiras linhas após renomear as colunas
aluguel_casa.head()
```

```python
# Verificação dos tipos de dados das colunas
aluguel_casa.dtypes
```

```python
# Conversão de tipos de dados para numérico (float)
aluguel_casa = aluguel_casa.astype({"rent_amount":"float", "property_tax":"float", "fire_insurance":"float", "total":"float"})
aluguel_casa.dtypes
```

```python
# Verificação dos tipos de dados após a conversão
aluguel_casa.head()
```

```python
# Exibição das primeiras linhas após a conversão de tipos
aluguel_casa.info()
```

```python
# Cálculo de estatísticas descritivas para colunas selecionadas
aluguel_casa[["area", "rooms", "bathroom", "parking_spaces", "rent_amount", "property_tax", "fire_insurance", "total"]].describe()
```

```python
# Filtragem de dados: preço total menor ou igual a 4000
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000)]
aluguel_casa.head()
```

```python
# Cálculo de estatísticas descritivas após a primeira filtragem
aluguel_casa.describe()
```

```python
# Filtragem de dados: preço total menor ou igual a 4000 e aceita animais
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept")]
aluguel_casa.head()
```

```python
# Filtragem de dados usando .query(): preço total <= 4000 e aceita animais
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept'")
aluguel_casa.head()
```

```python
# Filtragem de dados: preço total <= 4000, aceita animais e 2 ou 3 quartos
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept") & ((aluguel_casa["rooms"]==2)|(aluguel_casa["rooms"]==3))]
aluguel_casa.head()
```

```python
# Filtragem de dados usando .query(): preço total <= 4000, aceita animais e 2 ou 3 quartos
aluguel_casa = aluguel_casa.query(("total <= 4000 and animal == 'accept' and rooms == 2 or rooms == 3"))
aluguel_casa
```

```python
# Cálculo de estatísticas descritivas após todas as filtragens
aluguel_casa.describe()
```

```python
# Agrupamento por cidade e cálculo da média do preço total
agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
# Agrupamento por número de banheiros e quartos e cálculo da média do preço total
agg_bath_room_price = aluguel_casa.groupby(["bathroom","rooms"])["total"].agg("mean").reset_index()
agg_bath_room_price
```

```python
# Importação da biblioteca Seaborn para visualização de dados
import seaborn as sns
```

```python
# Criação de gráfico de barras: média do preço por número de quartos
ax = sns.barplot(data=agg_bath_room_price, x="rooms", y="total", color = "blue");
ax.set(xlabel = "qtd quartos", ylabel = "media do preço");
```

```python
# Criação de gráfico de barras: média do preço por número de banheiros
ax = sns.barplot(data=agg_bath_room_price, x="bathroom", y="total", color = "green");
ax.set(xlabel = "qtd banheiros", ylabel = "media do preço");
```
